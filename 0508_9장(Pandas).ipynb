```python
area = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기',
'강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']

pm25_avg = [20,13,15,20,19,16,14,18,20,14,19,23,23,14,15,13,14]
pm25_max = [22,18,18,30,23,20,16,19,27,18,26,29,30,27,23,17,16]
pm25_min = [16,10,11,12,15,12,11,17,13,9,14,19,14,10,8,9,11]

df = pd.DataFrame() # 빈 데이터프레임 생성
df['지역'] = area # ‘지역’ 열 생성
df['평균'] = pm25_avg # ‘평균’ 열 생성
df['최대'] = pm25_max # ‘최대’ 열 생성
df['최소'] = pm25_min # ‘최소’ 열 생성

df

print(df.loc[1,'지역'])
print(df.loc[1:3,'지역':'최대'])
print(df.loc[[1,3],['지역','최대']])
print("----------")
print(df.iloc[1,0])
print(df.iloc[1:4, 0:3])
print(df.iloc[[1,3], [0,2]])
print("----------")
print(df[df['지역']=='부산']) #부산과 관련된 행 전체 뽑아옴
print(df[df['최대']>=25]) #최대 값이 20 이상인 것
print(df[(df['최소']<=10)&(df['최대']>=25)]) #최소값이 10 이하이고 최대값이 25 이상인 것
```
