```python
#for문으로 세로 구구단 출력하기

for i in range(1,10):
    print("=%d단="%i, end=' ')
    print()
    for j in range(1,10):
        print("%d*%d=%d" %(i, j, i*j))
    print("---------")
```

```python
#for문으로 가로 구구단 출력하기
for i in range(1,10):
    print("=%2d단=" %i, end=' ')
print() 

for i in range(1,10):
    for j in range(1,10):
        print("%d*%d=%2d" %(j, i, i*j), end = ' ')
    print() #이렇게 해서 결국 줄바꿈 된 것 
```
