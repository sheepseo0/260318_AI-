```python
#함수 기본 (이것도 py파일에서 해보기)

def iplus(x1, x2): #iplus에 input이 2개 들어옴 (밑에서 4,5 줌. 어떤 걸 주냐에 따라 다름)
    isum = x1 + x2
    return isum # return : iplus라는 함수를 isum으로 바꿔주는 것 iplus(4,5) -> 9 이렇게!

hap = iplus(4,8) #이걸 위의 3중과 연관시키는 게 어렵다?

print(hap) #이게 최종 아웃풋!
```

```python
def root(a,b,c):
    x1 = (-b+(b**2.0-4.0*a*c)**0.5)/(2.0*a)
    x2 = (-b-(b**2.0-4.0*a*c)**0.5)/(2.0*a)
    return x1,x2

r1, r2 = root(1,2,3)

print(r1, r2)

#이렇게 하면 복소수로 나온다. 실수? 실수해가 있으면 코드 돌아가라....... 없으면 안 돌아가게 어쩌구....
```
