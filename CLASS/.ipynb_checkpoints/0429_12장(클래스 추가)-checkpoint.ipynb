```python
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return "({},{})".format(self.x, self.y)
    def __add__(self, other): #__ __ 하면 특수 함수가 된대요
        return Vector2D(self.x + other.x, self.y + other.y)
    def __sub__(self, other):
        return Vector2D(self.x - other.x, self.y - other.y)
    def __mul__(self, other):
        return self.x * other.x, self.y * other.y #벡터를 곱하면 숫자가 나온다 벡터가 나오는 게 아니라 
    def __gt__(self, other):
        return self.x**2+self.y**2 > other.x**2+other.y**2
    def __lt__(self, other):
        return self.x**2+self.y**2 < other.x**2+other.y**2
v1 = Vector2D(30, 40)
v2 = Vector2D(10, 20) #v2는 Vector2D로 만든 객체!
v3 = v1 + v2
print("v1 + v2 =",v3)
#v3 = v1.add(v2)
#print('v1.add(v2) =', v3)
iprod = v1*v2
print("v1*v2 =", iprod)

if(v1<v2):
    
if(v1>v2):
```
