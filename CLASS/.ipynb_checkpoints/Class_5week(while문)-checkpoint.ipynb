```python
isum = 0
n_list = [0,2,5,10]

for i in n_list:
    isum += 1

print(isum)

# = len(n_list) 이것과 같은 것
#만약에 합을 구하려면
'''
isum = 0
n_list = [0,2,5,10]

for i in n_list:
    isum += i

print(isum)
'''
```

```python
#이걸 while로 나타내면?

n_list = [0,2,5,10]
isum = 0
i = 0
while i < 4:#이렇게 하면 일일이 리스트의 갯수를 알아야함!!!
    isum += n_list[i]
    i += 1
print(isum)
```

```python
n_list = [0,2,5,10]
isum = 0
i = 0
while i < len(n_list): #그래서 이렇게 하면 됨
    isum += n_list[i]
    i += 1
print(isum)
```

```python
n = int(input("합계를 구할 수를 입력하세요 :"))
s = 0
i = 1
while i <= n:
    s = s + i
    i += 1
print("1부터 {}까지의 합은 {}".format(n,s))
```

```python
selected = None #selected란 변수를 None으로 처리함 (0으로 안 한 이유? 문자열이니까?)
while selected not in ['가위','바위','보']: #selected가 가위, 바위, 보 중에서 있는가? 없으면 실행을 안 함
    selected = input("가위, 바위, 보 중에서 선택하세요 >>>")
print("선택한 값은 :", selected)
```

```python
st = 'Programming'

#자음이 나타나는 동안만 출력하는 기능
for ch in st:
    if ch in ['a','e','i','o','u']:
        break #모음일 경우 반복문을 종료한다.
    print(ch)
print('The end')
```

```python
st = 'Programming'

#자음이 나타나는 동안만 출력하는 기능
for ch in st:
    if ch in ['a','e','i','o','u']:
        continue #모음일 경우 반복문을 종료한다.
    print(ch)
print('The end')
```
