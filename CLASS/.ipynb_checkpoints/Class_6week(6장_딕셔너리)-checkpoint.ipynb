```python
#6.4 딕셔너리의 메소드
person = {'이름':'홍길동','나이':26,'몸무게':82}
person['국적'] = "대한민국"

print(person.keys()) 
print(person.values())
print(person.items())
```

```python
person = {'이름':'홍길동','나이':26,'몸무게':82}
person['국적'] = "대한민국"

for key in person: #딕셔너리에서 key를 받아온 것? 
    print('{}:{}'.format(key, person[key])) #'{}:{}':플레이스 홀더 : (뒤에 있는 것)위치 고정헤주는 것
```

```python
def plusminus(a1, a2): #plusminus라는 함수 지정
    return a1+a2, a1-a2 #리턴해서 두 개를 받아왔다?(이게 뭔 말이지)

output = plusminus(10,2) #변경 불가능!

print(type(output)) 

#이렇게 하면 튜플로 나옴 왜야???
```

```python
#튜플 -> 리스트로 변환
def plusminus(a1, a2): 
    return a1+a2, a1-a2 

output = plusminus(10,2) 

output_list = list(output)
print(type(output_list)) 
```

```python
#swap : 바꾼다!
a = 100
b = 200
print("Before swap :", a, b)
a, b = b, a #뭐? 뒤에서 패킹을 해서 앞에서 언패킹을 해? (튜플 변수가 잇대)
print("After swap :", a, b)
```

```python
a = 100
b = 200
print("Before swap :", a, b)

a = b 
b = a 
#이케 하면 안됨.. a가 b가 됨.. 그럼 이젠 모든 게 a랑 같아지는거임. 그래서 a_org을 쓸
print("After swap :", a, b)
```

```python
a = 100
b = 200
print("Before swap :", a, b)
a_org = a
a = b 
b = a_org
print("After swap :", a, b)
#그럼 이젠 모든 게 a랑 같아지는거임. 그래서 a_org을 쓸??????
```

```python
#집합의 메소드
a_set = {1,3,5,6}
b_set = {1,2,4,6,7}

#c_set = a_set.union(b_set) #합집합 만들기

#c_set = a_set & b_set # 다 같은 것
c_set = a_set.intersection(b_set)
c_set = b_set & a_set
c_set = b_set.intersection(a_set)

c_set = a_set - b_set #이거랑
c_set = a_set.difference(b_set) #a를 기준으로 b를 빼는 것

c_set = b_set - a_set #이거는 다름
c_set = b_set.difference(a_set) 

print(c_set)
```
