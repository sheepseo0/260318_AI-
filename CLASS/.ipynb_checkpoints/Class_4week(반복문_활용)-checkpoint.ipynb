```python
s = 0
for i in range(1,11):
    s = s + i #루프 (계속 돈다)
    print(s)
    
print("1에서 10까지의 합:",s)
```

```python
#.<- 점이 들어가면 객체 지향? 뒤에가 메소드. 행동을 하는 것 (문자열에 행동이 들어간다!)
s = 0
for i in range(1,11):
    s = s + i 
    print('i = {}, s = {}'.format(i, s))
    
print("1에서 10까지의 합:",s)
```

```python
n = int(input("합계를 구할 수를 입력하시오.:"))
s = 0 

for i in range(0,n):
    s = s + (i+1) #n-1까지 생성하니까.. 조정해줘야겠지!

print("1부터 {}까지의 합은 {}".format(n, s))
```

```python
### 펙토리얼 계산하기
n = int(input("수를 입력하시오.:"))
fact = 1 #왜? 0부터 곱하면 0이 되니까
for i in range(1, n+1):
    fact = fact * i #숫자를 하나씩 곱해가는 것

print('{}! = {}'.format(n, fact))
```

```python
summer_fruits = ['수박','참외','체리','포도']

for fruit in summer_fruits: #for 변수 in 리스트 : 리스트에서 하나씩 꺼내어 변수에 넣고 반복하기
    print(fruit, end = ' ')
```

```python
#중첩 for 루프 : for문 안에 for문을 다시 넣음
for i in range(2,10):
    for j in range(1,10):
        print("{}*{} = {:2d}".format(i, j, i*j), end = ' ')
    print() #j 한바퀴 다 돌리고, print하고 그 다음 또 돌리고.. if, 이게 한 번 더 들여쓰기 되면 줄바꿈 됨 
```

```python
n = 7
for i in range(n):
    st = '' #st : 빈 문자열을 선언한 것
    for j in range(i):
        st = st+' ' # 공백 1개 추가함
    print(st +'*') # 공백 추가 후 # 출력
```

```python
n = 7

for i in range(n):
    print(' '*i+'♥') # 빈칸에 곱하기 한 것 (좀 더 효율적인 코드)
```

```python
# 소수 구하기 (중요함!)
n = int(input('수를 입력하세요:'))

is_prime = True
for num in range(2, n):
    if n%num == 0:
        is_prime = False
        break
print(n, 'is prime:', is_prime)

#한 번 False가 나오면 나와도 됨. 끝내야해. 어떻게? break!
#그리고 만약에 7을 입력해, 2,3,4로 다 나눠봤는데 안 나눠져. 그러면 그 보다 큰 수로는 나눌 필요가 없음 어차피 안 나눠지니까
```
