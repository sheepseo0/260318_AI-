```python
for _ in range(5): #언더스코어 : for i in range() 에서 변수 i는 실행문에서 사용되지 않는 루프변수라 언더스코어를 대신 넣어 익명화 시킬 수 있음
    print(i)
```

```python
for i in range(10):
    print(i+1, 'welcome to python!')
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(5):
    print(i, end =' ') #end가 없으면 기본으로 줄바꿈 기호가 들어간다. so, 이게 있어서 줄 바꿈 되지 X
```

```python
for i in range(5):
    print(i)
    print(i, end ='\n')
```

```python
for i in range(-2, -10, -2): #마지막을 양수로 주면 값이 안 나옴
    print(i, end = ' ')
```

```python
for i in range (0, 10, 2):
    print(i, end = ' ')
```
