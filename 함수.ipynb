```python
msg='''
1. 덧셈
2. 뺄셈
3. 곱셈
4. 나눗셈
5. 거듭제곱
6. 종료
'''
print(msg)

while True:
    menu = int(input("메뉴 선택하셈 : "))
    if menu == 1:
        print(f'덧셈 결과: {addition(input1, input2)}')
    elif menu == 2:
        print(f'뺄셈 결과: {subtraction(input1, input2)}')
    elif menu == 3:
        print(f'곱셈 결과: {multiplication(input1, input2)}')
    elif menu == 4:
        print(f'나눗셈 결과: {division(input1, input2)}')
    elif menu == 5:
        print(f'거듭제곱 결과: {power(input1, input2)}')
    else:
        print('프로그램을 종료합니다')
        break


def subtraction(num1, num2):
    if num1 > num2:
        return num1-num2
    else:
        return num2-num1
def multiplication(num1, num2):
    return num1 * num2
def division(num1, num2):
    if num2 == 0:
        return 0
    return num1 / num2
def power(num1, num2):
    result = 1
    for i in range(num2):
        result = multiplication(result, num1)
    return result
#숫자 입력
input1 = int(input('첫 번째 숫자를 입력하세요 >> '))
input2 = int(input('두 번째 숫자를 입력하세요 >> '))

    
```
