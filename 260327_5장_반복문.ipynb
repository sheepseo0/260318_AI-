```python
sum = 0
for i in range(1,101):
    sum+=i
print("1부터 100까지의 합은", sum, "입니다.")

'''
i   sum=sum+1
1   sum=0+1
2   sum=0+1+2
3   sum=0+1+2+3
4   sum=0+1+2+3+4
... 반복
'''
```

```python
#펙토리얼 계산하기 
n = int(input("숫자를 입력하시오.:"))

fact = 1
for i in range(1,n+1):
    fact *= i
print(n, "!은", fact, "이다.")
```
