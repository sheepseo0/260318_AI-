```python
import pandas as pd
df = pd.read_csv("weather1.csv", encoding='cp949')
df
```

```python
#1
df.describe()
```

```python
#2. 날짜에 대해 to_datetime으로 바꾸고 ‘년도’ 열 추가하기
df['날짜'] = pd.to_datetime(df['날짜']) #날짜 열 datetime 형식으로 바꿈
df['년도'] = df['날짜'].dt.year
df.head()
```

```python
#3. 몇 년도에 최대 기온이 나타나는가?
year_max = df.groupby('년도')['최고기온(℃)'].max()
print("최고 기온 발생 년도 :", year_max.idxmax())
print("최고 기온 :", year_max.max()) 
```

```python
#4. 몇 년도의 최소 기온이 가장 낮았는가?
year_min = df.groupby('년도')['최저기온(℃)'].min()

print("최소 기온 발생 년도 :", year_min.idxmin())  # 최저기온이 가장 낮았던 연도
print("최소 기온 :", year_min.min())
```

```python
#5. 평균 기온이 15 이상인 날 구하기
a = df[df['평균기온(℃)'] > 15]
a
```
